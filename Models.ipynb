```python
import numpy as np
import time as t
# import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report
import pickle
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LeakyReLU,Dropout
from tensorflow.keras.optimizers import SGD,Adam
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from tensorflow.keras.regularizers import l2,l1
```

```python
data_dict = pickle.load(open('./data_merged.pickle', 'rb'))
data = np.asarray(data_dict['data'])
labels = np.asarray(data_dict['labels'])
labels=labels.astype(int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, shuffle=True, stratify=labels)
```

```python
def evaluate_and_log(model_name, predict_fn,y=y_test):

    preds = predict_fn()
    accuracy = accuracy_score(y_test, preds)
    print(f"{model_name} accuracy = {accuracy:.4f}")
    print(f"{model_name} classification_report ={classification_report(y, preds)}")
```

```python
model1 = Sequential([
       
        Dense(256, activation='relu',input_shape=X_train[0].shape),
        Dense(128, activation='relu'),
        Dense(28, activation='softmax')
    ])

model1.compile(optimizer='adam',
                   loss='sparse_categorical_crossentropy',
                   metrics=['accuracy'])
start_time=t.time() 
model1.fit(X_train, y_train, epochs=5, batch_size=2048, validation_split=0.1)
end_time=t.time()
print(f"Time Taken for Training of DL_Model1={end_time-start_time}")



num_classes = 28

model2 = Sequential([
        Dense(128, activation='relu'),
        Dense(num_classes, activation='softmax')
    ])

model2.compile(optimizer='adam',
                   loss='sparse_categorical_crossentropy',
                   metrics=['accuracy'])
start_time=t.time() 
model2.fit(X_train, y_train, epochs=5, batch_size=2048, validation_split=0.1)
end_time=t.time()
print(f"Time Taken for Training of DL_Model2={end_time-start_time}")
```

```python
start_time=t.time() 
preds = np.argmax(model1.predict(X_test), axis=1)
evaluate_and_log("DL_Model_1", lambda: preds)
end_time=t.time()
print(f"Time Taken for Testing of DL_Model1={end_time-start_time}")
```

```python
start_time=t.time() 
preds = np.argmax(model2.predict(X_test), axis=1)
evaluate_and_log("DL_Model_2", lambda: preds)
end_time=t.time()
print(f"Time Taken for Testing of DL_Model2={end_time-start_time}")
```

```python
model3= Sequential()
print(f"{X_train.shape[1]}")
model3.add(Dense(100, activation='relu', input_shape=(X_train.shape[1],)))
model3.add(Dense(200, activation='relu'))
model3.add(Dropout(0.3))
model3.add(Dense(300, activation='relu'))
model3.add(Dense(150, activation='relu'))
model3.add(Dense(28, activation='softmax'))
optimi=Adam(learning_rate=0.005)
# early_stopping = EarlyStopping(monitor='val_loss', patience=3,min_delta=0.1, restore_best_weights=True)
model3.compile(optimizer=optimi, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
start_time=t.time() 
model3.fit(X_train, y_train, validation_split=0.1, epochs=200, batch_size=1024) 
end_time=t.time()
print(f"Time Taken for Training of DL_Model3={end_time-start_time}")
```

```python
start_time=t.time() 
y_pred=model3.predict(X_test).argmax(axis=1)
accuracy = accuracy_score(y_test, y_pred)  
print(f"The accuracy of the DL_Model_3={accuracy}")
print(classification_report(y_test, y_pred))
end_time=t.time()
print(f"Time Taken for Testing of DL_Model3={end_time-start_time}")
```
